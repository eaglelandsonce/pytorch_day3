## Lab 1: Use core tensor operations

### What you are learning
- What **in-place tensor ops** are (and why they can be risky if you need old values).
- How `torch.manual_seed()` supports **reproducibility** in experiments.


### 1) In-place operations like `add_()` modify the tensor directly in memory.  
A. True  
B. False  

<details>
<summary><strong>Answer + Why</strong></summary>

**Answer: A (True).**  
In-place ops (often ending with `_`) change the tensor’s contents without allocating new memory. This can be faster, but it can also cause bugs if you expected the original tensor to remain unchanged.

</details>


### 2) What does setting `torch.manual_seed()` help with?  
A. Enabling CUDA acceleration  
B. Tracking memory usage  
C. Ensuring reproducibility of random results  
D. Improving training speed  

<details>
<summary><strong>Answer + Why</strong></summary>

**Answer: C. Ensuring reproducibility of random results.**  
Seeding makes random number generation repeatable (to the extent possible), so you can re-run experiments and get consistent results.

</details>

## Lab 2: Backward graph visualization

### What you are learning
- How `.detach()` controls whether a tensor participates in the autograd graph.
- Why gradients **accumulate** unless you clear them.


### 1) What does the `.detach()` operation do?  
A. Computes gradients twice  
B. Multiplies the tensor by 2  
C. Removes the tensor from the computation graph  
D. Clears the gradients stored in `.grad`  

<details>
<summary><strong>Answer + Why</strong></summary>

**Answer: C. Removes the tensor from the computation graph.**  
Detached tensors don’t track gradients, so they won’t contribute to backpropagation.

</details>


### 2) Gradients are automatically cleared after each backward pass in PyTorch.  
A. True  
B. False  

<details>
<summary><strong>Answer + Why</strong></summary>

**Answer: B (False).**  
Gradients **accumulate by default** in PyTorch. You typically clear them each iteration with `optimizer.zero_grad()` or `param.grad.zero_()`.

</details>

## Lab 3: Create a custom MLP model

### What you are learning
- How `forward()` defines the computation a model performs.
- Why layers must be created in `__init__()` so parameters are registered once.


### 1) What is the role of the `forward()` method in an `nn.Module`?

<details>
<summary><strong>Answer + Why</strong></summary>

**Answer:** It defines how input data passes through the model layers.  
`forward()` is the blueprint for the computation graph when you call `model(x)`.

</details>


### 2) Why should layers be defined inside `__init__()`, not inside `forward()`?

<details>
<summary><strong>Answer + Why</strong></summary>

**Answer:** Defining layers in `__init__()` ensures their parameters are tracked and optimized correctly. Layers inside `forward()` would reinitialize on every pass.  
PyTorch can only optimize parameters that are registered as module attributes (created once).

</details>
